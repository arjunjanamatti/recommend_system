```python
mobile_df = pd.read_excel('../Data/mobile_phone_names.xlsx')
mobile_df.head(3)
```

```python
mobile_list = [_ for _ in mobile_df.iloc[:,0] if type(_) == str ]
mobile_list
```

```python
import re

mobile_list_updated = []
for string in mobile_list:
    mobile_list_updated.append(re.sub('<.*?>', '', string))
```

```python
mobile_list_updated
```

```python
new_df = pd.DataFrame()
new_df['model_name'] = mobile_list_updated
new_df
```

```python
new_df.to_csv('mobile_data.csv')
```
